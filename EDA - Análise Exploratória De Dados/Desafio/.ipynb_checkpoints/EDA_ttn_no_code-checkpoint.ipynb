## Imputa√ß√£o

```python
#Antes
ttn['Age'].describe()
```

```python
#Armazenando o valor desejado em uma vari√°vel: Mediana para os sobreviventes
male_surv_median = ttn[ttn['Survived'] == 1][ttn['Sex'] == 'male']['Age'].median()
#selecionar a coluna que vai receber os valores e especificar as condi√ß√µes = coluna.fillna(vari√°vel)
ttn['Age'][ttn['Survived'] == 1][ttn['Sex'] == 'male'] = ttn['Age'].fillna(male_surv_median);

female_surv_median = ttn[ttn['Survived'] == 1][ttn['Sex'] == 'female']['Age'].median()
ttn['Age'][ttn['Survived'] == 1][ttn['Sex'] == 'female'] = ttn['Age'].fillna(female_surv_median);

#Armazenando o valor desejado em uma vari√°vel: M√©dia para os n√£o sobreviventes
male_non_surv_mean = round(ttn[ttn['Survived'] == 0][ttn['Sex'] == 'male']['Age'].mean())
ttn['Age'][ttn['Survived'] == 0][ttn['Sex'] == 'male'] = ttn['Age'].fillna(male_non_surv_mean, inplace=True);

female_non_surv_mean = round(ttn[ttn['Survived'] == 0][ttn['Sex'] == 'female']['Age'].mean())
ttn['Age'][ttn['Survived'] == 0][ttn['Sex'] == 'female'] = ttn['Age'].fillna(female_non_surv_mean, inplace=True);
```
