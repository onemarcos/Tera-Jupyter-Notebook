```python
db = "chinook.db"

conn = sqlite3.connect(db)

cursor = conn.cursor()

query = '''SELECT DISTINCT billingcountry FROM invoices'''

cursor.execute(query)

result = pd.read_sql(query,conn)

result
```

```python
def run_query(q):
    with sqlite3.connect(db) as conn:
        return pd.read_sql(q, conn)


def show_tables():
    q = """
        SELECT
            name,
            type
        FROM sqlite_master
        WHERE type IN ("table");
        """
    return run_query(q)


show_tables()
```

### GROUP BY

```python
query = """SELECT albumid, COUNT(trackid) AS num_trackid
           FROM
                tracks
           GROUP BY
                albumid
           ORDER BY 
                num_trackid DESC;
"""

cursor.execute(query)

result = pd.read_sql(query, conn)

result
```

```python
query = """SELECT
                albumid,
                SUM(milliseconds) length,
                SUM(bytes) size
            FROM
                tracks
            GROUP BY
                albumid;"""

cursor.execute(query)

result = pd.read_sql(query, conn)

result

```

```python
query = """SELECT
                tracks.albumid,
                title,
                min(milliseconds),
                max(milliseconds),
                round(avg(milliseconds),2)
            FROM
                tracks
            INNER JOIN albums ON albums.albumid = tracks.albumid
            GROUP BY
                tracks.albumid;"""

cursor.execute(query)

result = pd.read_sql(query, conn)

result

```

## HAVING

```python
query = """ SELECT 
                albumid,
                COUNT(trackid)
            FROM
                tracks
            GROUP BY
                albumid
            HAVING albumid = 1;"""

cursor.execute(query)

result = pd.read_sql(query, conn)

result
```

```python
query = """ SELECT
               albumid,
               COUNT(trackid)
            FROM
               tracks
            GROUP BY
               albumid
            HAVING 
               COUNT(albumid) BETWEEN 18 AND 20
            ORDER BY albumid;"""

cursor.execute(query)

result = pd.read_sql(query, conn)

result
```
